```python
# Import Dependencies
import glob
import os
import pandas as pd   
```

### Join all the data files 

```python
# Read all the files and join them together
#https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe

Joined_citi_bike_data = pd.concat(map(pd.read_csv, glob.glob(os.path.join(".", "Resources", "*.csv"))))
Joined_citi_bike_data.shape
```

##### Note: Merged data has 336802 rows

```python
Joined_citi_bike_data
```

### Check for data types:

```python
# check for data types
Joined_citi_bike_data.dtypes
```

```python
Joined_citi_bike_data['starttime'] = pd.to_datetime(Joined_citi_bike_data['starttime'], errors='coerce')
Joined_citi_bike_data['stoptime'] = pd.to_datetime(Joined_citi_bike_data['stoptime'], errors='coerce')
#Joined_citi_bike_data['tripduration'] = pd.to_datetime(Joined_citi_bike_data['tripduration'], errors='coerce')
```

```python
# check for data types
Joined_citi_bike_data.dtypes
```

### Rename the columns:


```python
Joined_citi_bike_data.columns
```

```python
Joined_citi_bike_data = Joined_citi_bike_data.rename(columns={'tripduration':'Trip Duration', 'starttime':'Start Time', 
                                      'stoptime': 'Stop Time', 'start station id':'Start Station ID',
                                      'start station name':'Start Station Name', 'start station latitude':'Start Station Latitude',
                                      'start station longitude':'Start Station Longitude', 'end station id':'End Station ID', 
                                      'end station name':'End Station Name','end station latitude':'End Station Latitude', 
                                      'end station longitude':'End Station Longitude', 'bikeid':'Bike ID', 
                                      'usertype': 'User Type','birth year':'Birth Year', 'gender':'Gender'})
Joined_citi_bike_data
```

### Check to see if No trips made ie, start time equal to stop time

```python
Joined_citi_bike_data.loc[Joined_citi_bike_data['Start Time']== Joined_citi_bike_data['Stop Time'],:]
```

### Extract Months and days from start and stop

```python
from datetime import datetime

Joined_citi_bike_data['Month'] = pd.to_datetime(Joined_citi_bike_data['Start Time']).dt.month_name(locale = 'English')
Joined_citi_bike_data['Trip Week'] = Joined_citi_bike_data['Start Time'].dt.day_name(locale = 'English')
Joined_citi_bike_data
```
